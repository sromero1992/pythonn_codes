```python
# pip install matplotlib-venn

#Import libraries
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt 

dv = {"AA2", "AA3", "AB1", "AB2"} 
deup = {"AB1", "AB2"} 
dedn = {"AA2", "AA3"} 

dv_deup = dv.intersection(deup)
dv_dedn = dv.intersection(dedn)
deup_dedn = deup.intersection(dedn)
ovl_all = dv_dedn.intersection(dv_deup)

dv_tot  = dv.intersection(dv)- dv_deup - dv_dedn 
dedn_tot = dedn.intersection(dedn) 

dv_tot = 124
deup_tot = 167
dedn_tot = 157
deup_dedn = 1
dv_deup = 43
dv_dedn = 33
ovl_all = 1

# depict venn diagram 
v = venn3(subsets=(dv_tot, deup_tot, dedn_tot, dv_deup, dv_dedn, deup_dedn, ovl_all),  
      set_labels=('DV', 'DE up', 'DE dn'),  
      set_colors=("orange", "blue", "red"), alpha=0.7) 

# set color to defined path id 
#v.get_patch_by_id("100").set_color("white") 
# set text to defined label id 
#v.get_label_by_id("100").set_text("unknown") 
# set text to defined label id "A" 
#v.get_label_by_id('A').set_text('A new') 

# outline of circle line style and width 
venn3_circles(subsets=(dv_tot, deup_tot, dedn_tot, dv_deup, dv_dedn, deup_dedn, ovl_all), 
              linestyle="dashed", linewidth=2) 
  

   
# title of the venn diagram 
plt.title("Venn Diagram for methods") 
plt.show() 
```
